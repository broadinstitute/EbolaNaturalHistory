```r
# read in lab master sheet #

master_table<-read.csv2("qPCR_table.csv",sep=",")
#master_table<-read.csv2("Master_Sample-Keys_TotalRNA_EBOV-201028.tsv",sep="\t")

# make some corrections to master table #
master_table$ID.Individual<-gsub("RA0552","RA0522",master_table$ID.Individual) # typo
master_table$ID.Cohort<-gsub("[?]","",master_table$ID.Cohort) # typo
master_table$ID.Cohort<-gsub("D-28","D-28/30",master_table$ID.Cohort) # merge D28 and D30
master_table$ID.Cohort<-gsub("D-30","D-28/30",master_table$ID.Cohort) # merge D28 and D30
master_table$ID.Cohort<-gsub("D-4","D-4/14",master_table$ID.Cohort) # merge D28 and D30
master_table$ID.Cohort<-gsub("D-04","D-4/14",master_table$ID.Cohort) # merge D28 and D30
master_table$ID.Cohort<-gsub("D-14","D-4/14",master_table$ID.Cohort) # merge D28 and D30
master_table$ID.Cohort<-gsub("D-4/14","<0",master_table$ID.Cohort) # merge D28 and D30
master_table$ID.Cohort<-gsub("D-28/30","<0",master_table$ID.Cohort) # merge D28 and D30
unique(master_table$ID.Cohort)

# swap 1819 and 1074 tissues #
solid_tissues<-c('Spleen','Kidney','Liver','Adrenal Gland','Brain','Sex Organ','Lymph node','Lung','Skin')
fluids<-c('Whole blood','CSF','Serum','PBMC')
#should_be_1074<-which(master_table$ID.Individual=="RA1819" & master_table$Tissue %in% solid_tissues)
#should_be_1819<-which(master_table$ID.Individual=="RA1074" & master_table$Tissue %in% solid_tissues)

master_table$ID.Individual<-as.character(master_table$ID.Individual)
master_table$ID.Cohort<-as.character(master_table$ID.Cohort)
#master_table$ID.Individual[should_be_1074]<-"RA1074"
#master_table$ID.Cohort[should_be_1074]<-"D006"

#master_table$ID.Individual[should_be_1819]<-"RA1819"
#master_table$ID.Cohort[should_be_1819]<-"D000"

# make a more fine-grained tissue variable #
master_table$Tissue[master_table$Tissue=="CSF" & master_table$ID.Cohort=="NaN"]<-"CSF-BL"

tissue_resolved<-sapply(master_table$Biosample,function(x) strsplit(x,"_")[[1]][2]) # remove individual ID
tissue_resolved<-sapply(tissue_resolved,function(x) strsplit(x,"-D")[[1]][1]) # remove day or other details
tissue_resolved<-gsub("Ovary","Ovary/testis",tissue_resolved)
tissue_resolved<-gsub("Testis","Ovary/testis",tissue_resolved)
tissue_resolved<-gsub("Liver-L","Liver",tissue_resolved)
tissue_resolved<-gsub("LNG-CD-R","Lung",tissue_resolved)
master_table$TissueDetail<-tissue_resolved
                        
head(master_table)
names(master_table)
master_table_WB=master_table
```

```r
### Correct Metadata
suppressPackageStartupMessages(library(data.table))
suppressPackageStartupMessages(library(tidyverse))
source('../prj_helpers.R')
master_table_WB<-master_table
master_table_WB$sub.tissue <- c("")
master_table_WB[master_table_WB$Tissue == 'Skin',]$sub.tissue <-
  ifelse(grepl("Non-Rash",master_table_WB[master_table_WB$Tissue == 'Skin',]$Biosample), "Non-Rash", "Rash")

master_table_WB$sub.tissue

master_table_WB[master_table_WB$Tissue == 'Brain',]$sub.tissue <-
  ifelse(grepl("Brain-Gr",master_table_WB[master_table_WB$Tissue == 'Brain',]$Biosample), "Brain-Gr", "Brain-Wh")


sel.ln <- master_table_WB$Tissue == 'Sex Organ'
master_table_WB$sub.tissue[sel.ln] <- strpart(master_table_WB$Biosample[sel.ln],'_',2)


sel.ln <- master_table_WB$Tissue == 'Lymph node'
master_table_WB$sub.tissue[sel.ln] <- strpart(master_table_WB$Biosample[sel.ln],'_',2)
master_table_WB$sub.tissue[sel.ln]


master_table_WB$full.tissue <- paste0(master_table_WB$Tissue,':',master_table_WB$sub.tissue)
master_table_WB$full.tissue <- gsub(':$','',master_table_WB$full.tissue)
master_table_WB$full.tissue <- as.factor(master_table_WB$full.tissue )
```

```r
master_table_WB$quant.kulesh<-log10(as.numeric(master_table_WB$Quant.Kulesh)+1)
master_table_WB$dpi_time <- as.numeric(gsub('^D','',master_table_WB$ID.Cohort))
melt.logVirCounts<-master_table_WB[,c("Biosample","dpi_time","full.tissue","quant.kulesh")]
melt.logVirCounts
melt.logVirCounts$full.tissue[melt.logVirCounts$full.tissue %in%
                             c("Sex Organ:Ovary","Sex Organ:Sex-Organ","Sex Organ:Testis")]="Sex Organ:Sex-Organ"

mean.log.virus<-melt.logVirCounts %>%
  group_by(full.tissue, dpi_time) %>%
  summarise(mean=(mean(quant.kulesh,na.rm = TRUE)))
dataCluster<-mean.log.virus %>% filter(!full.tissue %in% c("PBMC"))%>%filter(!dpi_time %in% c(1,2,NA))
dim(dataCluster)

temp<-read.csv(file = "../01-data/full_tissue_color.csv",header = F)
color_ft<-temp$V2
names(color_ft)<-temp$V1
color_ft["Serum"]<-"#69dead"
color_ft["CSF"]<-"#0c3819"
color_ft
```

```r
# plot average counts by time #
library(kml)
#dataCluster%>%filter(variable %in% c("quant.kulesh"))
data_wide <- dataCluster %>% 
  dplyr::arrange(dpi_time) %>% 
  tidyr::pivot_wider(id_cols = full.tissue, names_from = dpi_time, values_from = mean)

head(data_wide)

data_cld <- kml::cld(as.data.frame(data_wide), timeInData = 2:8, maxNA = 1)
kml::kml(data_cld, nbClusters=2:3,nbRedrawing = 2,toPlot = 'none')

plot(data_cld, 3, parTraj=parTRAJ(col="clusters"), toPlot = "traj",)

clusters<-getClusters(data_cld,3,1)
clusters<-as.character(clusters)
clusters[clusters=="A"]<-"Cluster 3"
clusters[clusters=="B"]<-"Cluster 2"
clusters[clusters=="C"]<-"Cluster 1"
names(clusters)<-data_wide$full.tissue
data_wide["Cluster"]<-clusters
data.quant<-melt(data_wide)


ggplot(data.quant,aes(x = variable,y=value,col=full.tissue))+theme_cowplot()+
geom_point()+geom_line(aes(group=full.tissue))+ 
facet_wrap(~ Cluster,scales = "free_y")+
scale_color_manual(name="Tissue",values =color_ft[unique(dataCluster$full.tissue)])+
ggtitle("qPCR")
clusters
melt.logVirCounts.quant<-melt.logVirCounts %>% filter(full.tissue %in% unique(data.quant$full.tissue))
melt.logVirCounts.quant["cluster"]<-clusters[as.character(melt.logVirCounts.quant$full.tissue)]

ggplot(melt.logVirCounts.quant,aes(x = dpi_time,y=quant.kulesh,col=full.tissue))+theme_cowplot()+
geom_point(size=1)+geom_smooth(se = TRUE,scale="free_y",method = "loess",span=1,alpha=0.1,size=0.5)+
ggtitle("")+xlab("DPI")+ylab("qPCR")+
facet_wrap(~ cluster)+
    theme(text=element_text(size=10,color='black'))+
scale_color_manual(name="Tissue",values =color_ft[unique(as.character(melt.logVirCounts.quant$full.tissue))],guide="none" )
ggsave("../cluster_qpcr.pdf",width=8,height=3)
#ggplot(melt.logVirCounts.quant,aes(x = dpi_time,y=quant.kulesh,col=full.tissue))+theme_cowplot()+
#geom_point()+geom_smooth(method = "lm", formula = y ~ x + I(x^2), alpha = .15, se=T)+
#ggtitle("Reads")+
#facet_wrap(~ cluster,scales = "free")+scale_color_manual(name="Tissue",values =color_ft[unique(as.character(melt.logVirCounts.quant$full.tissue))] )


  
```
